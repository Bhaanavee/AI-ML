```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/Users/bhaanaveecs/Documents/ML_I/Titanic-Dataset.csv')
df.sample(5)
```

```python
numerical_features=df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(10,8))
for i,feature in enumerate(numerical_features):
    plt.subplot(3,3,i+1)
    sns.histplot(df[feature],kde=True, bins = 30, color='blue')
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
for i,feature in enumerate(numerical_features):
    plt.subplot(3,3,i+1)
    sns.boxplot(x=df[feature],color='green')
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
correlation_matrix=df[numerical_features].corr()
```

```python

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
nf=list(numerical_features)
if 'PassengerId' in nf:
    nf.remove('PassengerId')
plt.figure(figsize=(8,8))
sns.pairplot(df[nf])
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
fig = px.scatter(df.dropna(subset=['Age', 'Fare']), x='Age', y='Fare', color='Survived', title='Fare vs Age (Colored by Survival)')
fig.show()
```

```python
#Analysis
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Sex')
plt.show()

sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival Count by Embarked Port')
plt.show()
```

```python
'''Conclusions -
1.Women had a higher survival rate than men.
2.Higher Fare often correlated with higher survival chances.
3.Embarked location may influence survival probability.
4.Age distribution among survivors shows more survival in younger passengers.'''
```
